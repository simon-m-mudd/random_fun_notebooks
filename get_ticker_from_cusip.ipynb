```python
# prompt: I have a bunch of csv files in a directory that follow the file naming pattern screened*.csv
# I want to read these files and combine them into a single pandas dataframe, and print this to csv

import pandas as pd
import glob

# Get a list of all CSV files matching the pattern
csv_files = glob.glob("screened*.csv")

# Create an empty list to store DataFrames
df_list = []

# Loop through the list of CSV files
for file in csv_files:
  # Read each CSV file into a DataFrame
  df = pd.read_csv(file)
  # Append the DataFrame to the list
  df_list.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Print the combined DataFrame to a CSV file
combined_df.to_csv("combined_screened_data.csv", index=False)
```
